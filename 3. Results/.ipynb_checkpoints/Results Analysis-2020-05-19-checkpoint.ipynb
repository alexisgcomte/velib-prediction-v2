# Results Analysis - 2020-05-19

## Introduction

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
# Loading the dataset
df = pd.read_csv('Consolidated_results_all_models_2020-05-19.csv')
df.head()
```

## I. First Analysis

```python
# Analysis for global RMSE by model
df.groupby('model_name').metrics.mean()
```

LSTM_A_Global and LSTM_B_Global seem to work best

```python
# Analysis for global RMSE by hour
print(df.groupby('Date').metrics.mean())
```

```python
# Analysis for global RMSE by station
print(df.groupby('station').metrics.mean())
```

## II. Analysis by date

We will find out which models is the "wrongest", ie wich model have a max RMSE by Date over all the station.

```python
# Max RMSE by station and time
pd.pivot_table(df, values='metrics', columns='model_name', index='Date',  aggfunc='max')
```

```python
# Average max RMSE by date and model
print('Average max RMSE by date and model')

print('LSTM_A_Global : ',round(pd.pivot_table(df, values='metrics', columns='model_name', index='Date',  aggfunc='max')['LSTM_A_Global'].mean(),2))
print('LSTM_B_Global : ',round(pd.pivot_table(df, values='metrics', columns='model_name', index='Date',  aggfunc='max')['LSTM_B_Global'].mean(),2))
print('LSTM_A_By_Station : ',round(pd.pivot_table(df, values='metrics', columns='model_name', index='Date',  aggfunc='max')['LSTM_A_By_Station'].mean(),2))
print('LSTM_B_By_Station : ',round(pd.pivot_table(df, values='metrics', columns='model_name', index='Date',  aggfunc='max')['LSTM_B_By_Station'].mean(),2))
```

LSTM_B_Global is the least extreme, followed bu LSTM_A_Global.

## III. Analysis by station

We will find out which models is the "wrongest", ie wich model have a max RMSE by station over all the dates.

```python
# Worst predicted station
pd.pivot_table(df, values='metrics' ,index='station',  aggfunc='max').sort_values(by='metrics', ascending=False).iloc[:20,:]
```

```python
# Let's create a dataframe to analysis by station the max RMSE over a day

df_by_station = pd.pivot_table(df, values='metrics', columns='model_name', index='station',  aggfunc='max')
df_by_station.head(30)
```

```python
# Adding best model and best metrics for further analysis 

df_by_station["best_model"] = "" 
df_by_station["best_metrics"] = "" 
for i in df_by_station.index:
    if min(df_by_station.loc[i][:4]) == df_by_station["LSTM_A_Global"].loc[i]:
        df_by_station["best_model"].loc[i] = "LSTM_A_Global"
        df_by_station["best_metrics"].loc[i] = df_by_station["LSTM_A_Global"].loc[i]
    elif min(df_by_station.loc[i][:4]) == df_by_station["LSTM_B_Global"].loc[i]:
        df_by_station["best_model"].loc[i] = "LSTM_B_Global"
        df_by_station["best_metrics"].loc[i] = df_by_station["LSTM_B_Global"].loc[i]
    elif min(df_by_station.loc[i][:4]) == df_by_station["LSTM_A_By_Station"].loc[i]:
        df_by_station["best_model"].loc[i] = "LSTM_A_By_Station"
        df_by_station["best_metrics"].loc[i] = df_by_station["LSTM_A_By_Station"].loc[i]
    else:
        df_by_station["best_model"].loc[i] = "LSTM_B_By_Station"
        df_by_station["best_metrics"].loc[i] = df_by_station["LSTM_B_By_Station"].loc[i]
        
df_by_station.head(20)
```

```python
# Which model works best? Let's analysis which model wins in proportion: the minimum of max averaged RMSE by date
summary_table = pd.pivot_table(df_by_station, values='best_metrics', index='best_model',  aggfunc='count')
summary_table["%"] = ""
for i in summary_table.index:
    summary_table["%"][i] = str(round(summary_table.best_metrics[i] / summary_table.best_metrics.sum()*100))
print(summary_table.sort_values(by='%', ascending=False)['%'])
```

LSTM_A_Global and LSTM_B_Global seem to work well in average

```python
# Average max RMSE by station and model
print('Average max RMSE by station and model')

print('LSTM_A_Global : ',round(pd.pivot_table(df, values='metrics', columns='model_name', index='station',  aggfunc='max')['LSTM_A_Global'].mean(),2))
print('LSTM_B_Global : ',round(pd.pivot_table(df, values='metrics', columns='model_name', index='station',  aggfunc='max')['LSTM_B_Global'].mean(),2))
print('LSTM_A_By_Station : ',round(pd.pivot_table(df, values='metrics', columns='model_name', index='station',  aggfunc='max')['LSTM_A_By_Station'].mean(),2))
print('LSTM_B_By_Station : ',round(pd.pivot_table(df, values='metrics', columns='model_name', index='station',  aggfunc='max')['LSTM_B_By_Station'].mean(),2))
```

Here again, LSTM_A_Global and LSTM_B_Global seem to be the best models
