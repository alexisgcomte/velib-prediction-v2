## II. MySQL Creation of the tables

### Creation of the database
CREATE DATABASE db_velib;

### Creation of the table
CREATE TABLE velib_scrapping (
id INT UNSIGNED NOT NULL AUTO_INCREMENT ,
station_id SMALLINT UNSIGNED,
station_name VARCHAR(30),
operational_station VARCHAR(3),
station_capacity TINYINT UNSIGNED,
nb_free_docks TINYINT UNSIGNED,
nb_total_free_bikes TINYINT UNSIGNED,
nb_free_mechanical_bikes TINYINT UNSIGNED,
nb_free_electrical_bikes TINYINT UNSIGNED,
payment_totem VARCHAR(3),
bike_return_possible VARCHAR(3),
data_actualisation VARCHAR(30),
geo_coordinates VARCHAR(30),
city VARCHAR(30),
insee_code VARCHAR(15),
PRIMARY KEY (id)
) 
ENGINE=INNODB;

```python
# SECOND TABLE LIGHTER

### Creation of the table
CREATE TABLE velib_realtime (
station_id SMALLINT UNSIGNED,
operational_station VARCHAR(3),
nb_free_docks TINYINT UNSIGNED,
nb_total_free_bikes TINYINT UNSIGNED,
nb_free_mechanical_bikes TINYINT UNSIGNED,
nb_free_electrical_bikes TINYINT UNSIGNED,
payment_totem VARCHAR(3),
bike_return_possible VARCHAR(3),
data_actualisation DATETIME, # in minutes
date_of_update DATETIME,
PRIMARY KEY (station_id, date_of_update)
) 
ENGINE=INNODB;
```

```python
# 3rd TABLE for JSON file

### Creation of the table
CREATE TABLE velib_realtime_json (
station_id SMALLINT UNSIGNED,
num_bikes_available TINYINT UNSIGNED,
num_docks_available TINYINT UNSIGNED,
is_installed TINYINT UNSIGNED,
is_returning TINYINT UNSIGNED,
is_renting TINYINT UNSIGNED,
last_reported DATETIME
num_bike_aviailable_mechanical TINYINT UNSIGNED,
num_bike_aviailable_ebike TINYINT UNSIGNED,
date_of_update DATETIME,
PRIMARY KEY (station_id, date_of_update)
) 
ENGINE=INNODB;
    
```

```python
# temporary TABLE for JSON file


### Creation of the table
CREATE TABLE temp_velib_realtime_json (
stationCode SMALLINT UNSIGNED,
station_id SMALLINT UNSIGNED,
num_bikes_available SMALLINT UNSIGNED,
numBikesAvailable SMALLINT UNSIGNED,
num_docks_available SMALLINT UNSIGNED,
numDocksAvailable SMALLINT UNSIGNED,
is_installed SMALLINT UNSIGNED,
is_returning SMALLINT UNSIGNED,
is_renting SMALLINT UNSIGNED,
last_reported SMALLINT UNSIGNED,
num_bike_aviailable_mechanical SMALLINT UNSIGNED,
num_bike_aviailable_ebike SMALLINT UNSIGNED
) 
ENGINE=INNODB;
```
